#### Data Manipulation and Analysis with Pandas
Data manipulation and analysis are key tasks in any data science or data analysis project. Pandas provides a wide range of functions for data manipulation and analysis, making it easier to clean, transform, and extract insights from data. In this lesson, we will cover various data manipulation and analysis techniques using Pandas.

```python
df_filled=df.fillna(0)
print(df_filled)
```

```python
### filling missing values with the mean of the column
df['Sales_fillNA']=df['Sales'].fillna(df['Sales'].mean())
df
```

```python
df.dtypes
```

```python
## Renaming Columns
df=df.rename(columns={'Date':'Sales_Date','Sales':"Total_sales"})
df.head()
```

```python
## change datatypes
df['Value_new']=df['Value'].fillna(df['Value'].mean()).astype(int)
df.head()
```

```python
df=df.drop(columns='Value_new')
df.head()
```

```python
df['Value_new']=df['Value'].fillna(df['Value'].mean()).astype(int)
df.head()
```

```python
df['applied_func']=df['Value'].apply(lambda x:x**3)
df.head()
```

```python
df['New Value']=df['Value'].apply(lambda x:x*2)
df.head()
```

```python
## Data Aggregating And Grouping
df.head()
```

```python
grouped_mean=df.groupby('Product')['Value'].mean()
print(grouped_mean)
```

```python
grouped_sum=df.groupby(['Product','Region'])['Value'].sum()
print(grouped_sum)
```

```python
df.groupby(['Product','Region'])['Value'].mean()
```

```python
## aggregate multiple functions
groudped_agg=df.groupby('Region')['Value'].agg(['mean','sum','count'])
groudped_agg
```

```python
### Merging and joining Dataframes
# Create sample DataFrames
df1 = pd.DataFrame({'Key': ['A', 'B', 'C'], 'Value1': [1, 2, 3]})
df2 = pd.DataFrame({'Key': ['A', 'B', 'D'], 'Value2': [4, 5, 6]})
```

```python
df1
```

```python
df2
```

```python
## Merge Datafranme on the 'Key Columns'
pd.merge(df1,df2,on="Key",how="inner")
```

```python
pd.merge(df1,df2,on="Key",how="outer")
```

```python
pd.merge(df1,df2,on="Key",how="left")
```

```python
pd.merge(df1,df2,on="Key",how="right")
```

```python
# coverting date column dtype to datetime from object
df['Date']=pd.to_datetime(df['Date'])
```

```python
# changing the product category column to category tyype
df['Product Category']=df['Product Category'].astype('category')
```

```python
df.dtypes
```

```python
# grouping by product category to find the total no.of units sold and revenue generated
df.groupby('Product Category')[['Units Sold','Total Revenue']].sum()
```

```python
df["Product Category"].unique
```

```python
# grouping by both Product Category and region
df.groupby(['Product Category','Region'],observed=False)[['Units Sold','Total Revenue']].sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=df.groupby(['Product Category','Region'],observed=False)[['Units Sold','Total Revenue']].sum(), x='Product Category', y='Units Sold', hue='Region')

plt.title("Units Sold by Product Category and Region")
plt.xlabel("Product Category")
plt.ylabel("Units Sold")
plt.legend(title="Region")
plt.show()
```
